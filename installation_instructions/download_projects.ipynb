# Project Installation Instructions

open a terminal or powershell window and navigate to the `cs220` folder.  If you do not remember the commands or do not have the folder setup, please follow the [software installation instructions](https://cdis.pages.doit.wisc.edu/cs/courses/cs220/cs220-lecture-material/installation_instructions/) for setting this up.

Run the pixi command:

```
pixi run get_project <project>
```

Replacing `<project>` with which project it is (e.g. p1, p2, p3, etc.).  This should download and unzip the project file and setup a folder inside the `cs220/projects` directory.

To start Jupyter lab to begin working on the project type the pixi command:

```
pixi run jupyterlab
```

If you are still struggling, come to the [Gross Learning Center](https://cdis.wisc.edu/learning-center/) on the 2nd floor of Morgridge hall during drop-in hours to get some help.
