## Use pre-written Python code to read a CSV file into a list of lists.

```python


# inspired by https://automatetheboringstuff.com/chapter14/
import csv

def process_csv(filename):
    # open the file, its a text file utf-8
    exampleFile = open(filename, encoding="utf-8")  
    
    # prepare it for reading as a CSV object
    exampleReader = csv.reader(exampleFile) 
    
    # use the built-in list function to convert this into a list of lists
    exampleData = list(exampleReader)        
    
    # close the file to tidy up our workspace
    exampleFile.close()  
    
    # return the list of lists
    return exampleData
```

```python
# Call the process_csv function and store the list of lists in cs220_csv
cs220_csv = process_csv('amfam_survey_data.csv')
cs220_csv[:3]
```

```python
# Store the header row into cs220_header
cs220_header = cs220_csv[0]
cs220_header
```

```python
# Store all of the data rows into cs220_data
cs220_data = cs220_csv[1:]
cs220_data
```

```python
# Print out every student's sleep habit and pets
# use for 
for row in cs220_data:
    current_sleep_habit = row[cs220_header.index("sleep preference")]
    pets = row[cs220_header.index("pets")]
    print(current_sleep_habit + '\t\t' + pets)
```

```python
# Find the average years of people who do not live in Wisconsin
total_years = 0
count = 0
for row in cs220_data:
    years = float(row[cs220_header.index("years")])
    state = (row[cs220_header.index("state")])
    if state not in ["Wisconsin", "WI", "madison"]:
        total_years += years
        count += 1
        print(state, years)
avg = total_years / count
avg
```
