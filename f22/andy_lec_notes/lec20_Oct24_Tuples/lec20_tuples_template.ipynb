```python
# example 5d:  try in Python Tutor
def first(items):
    return items[0]

def smallest(items):
    items.sort()
    return items[0]

numbers	= [4,5,3,2,1]
print("first:", first(numbers))
print("smallest:", smallest(numbers))
print("first:", first(numbers))

# takeaway:  what happens when a list parameter is sorted "in place" using .sort() ?
# your answer: 
```
