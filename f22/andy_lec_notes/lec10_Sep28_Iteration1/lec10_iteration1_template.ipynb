### for loop

- does not require initialization of loop control variable outside the loop
- loop statement itself creates the loop control variable
- keywords `for` and `in`

### range() function
- range(5) produces the sequence 0, 1, 2, 3, 4, 
- range(3,5) produces the sequence 3, 4
- the ending number is not included


```python
# Example 5:  Another kind of a loop is a for loop
# the loop control variable in this loop is increased by 1 in each iteration

for i in range(3,7):
    print(i) # don't need to change i
```
