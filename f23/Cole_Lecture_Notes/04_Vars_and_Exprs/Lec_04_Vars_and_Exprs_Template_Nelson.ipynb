## Reminder...

- An **operator** is either mathematical (*, /, +, etc.,), comparison (==, <, etc.,), or logical (`not`, `and`, `or`).
- An **operand** is the data value that an operator operates on.
    - **unary operators** of signs are applied on a single operand. 
    - **binary operators** are applied on two operands.
- An **expression** is the combination of **operator(s)** and **operand(s)**.

## Short Circuiting

- Logical operators are read left-to-right
- If one of the expressions in an OR statement is `True`, the entire statement is `True`.
- Likewise, if one of the expressions in an AND statement is `False`, the entire statement is `False`.
- Don't waste time evaluating!

```python
# What will this expression evaluate to?
1 + 1 == 2 or 1 / 0 == -100
```

```python
# What will this expression evaluate to?
0 + 0 == 1 and 1 / 0 == -100
```

```python
# What will this expression evaluate to?
1 / 0 == -100 and 0 + 0 == 1
```

## Correct way of writing Boolean expressions

- Logical operators (`and`, `or`, `not`) should only accept booleans!

```python
# How to say is 6 greater than 1 and 2?
print(6 > 1 and 2) # What?
print(6 > 1 and 6 > 2) # Correct
```

## Variables
`10`, `22.0`, `"hello"`, `'goodbye'`, `True`, `False`... are known as *literal values*.

However, variables *vary* over time. They are assigned using the `=` operator. The left hand side must be a variable, and the right hand side must be an expression!

Variables look like... `x`, `player_score`, `my_cool_name`, `some_thing`... You can name them almost whatever you like (see slides for details)

```python
# We create a variable called "x" and assign it the value 1.5.
x = 1.5

# We create a new variable called "y" and assign the value of x * 3 to it.
y = x * 3

# We print the values of x, y, and x + y. What do you think these will be?
print(x)
print(y)
print(x + y)
```

```python
# Notice how we give meaningful variable names!
# It would be confusing to have variable names such as v, w, x, y, and z.

percent_burn = 0.25
sheets_of_cookie_dough = 5

all_cookies = sheets_of_cookie_dough * 12
burned_cookies = all_cookies * percent_burn
good_cookies = all_cookies - burned_cookies

print('All Cookies')
print(all_cookies)
print()

print('Burned Cookies')
print(burned_cookies)
print()

print('Good Cookies')
print(good_cookies)
print()
```

```python
# Using this len function, we can align strings more easily!
player1_name = 'Alice'
player2_name = 'bob'

player1_score = 10
player2_score = 8

#Scores are difficult to compare
print(player1_name + ': ' + '|' * player1_score)
print(player2_name + ': ' + '|' * player2_score)

#Assuming name has a maximum of 9 characters
#Now scores are easier to read and compare
print(player1_name + ': ' + " " * (9 - len(player1_name)) + '|' * player1_score)
print(player2_name + ': ' + " " * (9 - len(player2_name)) + '|' * player2_score)
```
